```python
with open('data/countries.geojason') as f:
    geojson_data = json.load(f)

new_df = pd.json_normalize(geojson_data)
new_df.head()
```

```python
# Extract properties from each feature
properties_list = [feature ['properties']for feature in geojson_data['features']]

# Convert the list of properties into a DataFrame
properties_df = pd.DataFrame(properties_list)

# Display the DataFrame
properties_df.head()
```

```python
# Extract properties from each feature
geometry_list = [feature['geometry'] for feature in geojson_data['features']]

# Convert the list of properties into a DataFrame
geometry_df = pd.DataFrame(geometry_list)

# Display the DataFrame
geometry_df.head()
```

```python
# merging
merged_df = pd.merge(properties_df,geometry_df,left_index=True, right_index=True)
merged_df.head()
```

```python
final_df = merged_df[['sovereignt','coordinates']]
final_df.head()
```

```python
final_df = final_df.rename(columns={
    'sovereignt': 'Country'	
})
final_df.head()
```

```python
#merging cost index df and the final df
data_df = pd.merge(Cost_df,final_df, how= "inner", on = 'Country')
data_df.head()
```

```python
data_df.to_json('livingcostbyCountry.json')
```
